```python
# #Scrape the details of most viewed videos on YouTube from Wikipedia. Url 
# = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
# Rank 
# B) Name 
# C) Artist 
# D) Upload date 
# E) Views

#
#pip install requests BeautifulSoup4

import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

# Send a GET request to fetch the HTML content
response = requests.get(url)

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the data of interest
table = soup.find("table", {"class": "wikitable"})

# Initialize lists to store data
rankings = []
names = []
artists = []
upload_dates = []
views = []

# Iterate through each row (skip header row)
for row in table.find_all("tr")[1:]:
    columns = row.find_all("td")
    if len(columns) >= 6:
        # Extract data from columns
        rank = columns[0].text.strip()
        name = columns[1].text.strip()
        artist = columns[2].text.strip()
        upload_date = columns[3].text.strip()
        view_count = columns[4].text.strip()

        # Append data to respective lists
        rankings.append(rank)
        names.append(name)
        artists.append(artist)
        upload_dates.append(upload_date)
        views.append(view_count)

# Print the scraped data (you can modify this to save to a file or use as needed)
for i in range(len(rankings)):
    print(f"Rank: {rankings[i]}")
    print(f"Name: {names[i]}")
    print(f"Artist: {artists[i]}")
    print(f"Upload Date: {upload_dates[i]}")
    print(f"Views: {views[i]}")
    print()

```

```python
# 3. Scrape the details of State-wise GDP of India from statisticstime.com. 
# Url = http://statisticstimes.com/ 
# You have to find following details: A) Rank 
# B) State 
# C) GSDP(18-19)- at current prices 
# D) GSDP(19-20)- at current prices 
# E) Share(18-19) 
# F) GDP($ billion) 
# Note: - From statisticstimes home page you have to reach to economy page through code.


import requests
from bs4 import BeautifulSoup

# Function to get the URL of economy page from Statistics Times homepage
def get_economy_url():
    url = "http://statisticstimes.com/"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            # Find the link to the economy page
            economy_link = soup.find("a", text="Economy")
            if economy_link:
                return economy_link["href"]
            else:
                print("Economy link not found.")
        else:
            print(f"Failed to fetch homepage. Status code: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching homepage: {str(e)}")

    return None

# Function to scrape state-wise GDP data
def scrape_gdp_data(economy_url):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }
        response = requests.get(economy_url, headers=headers)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            
            # Find the table containing state-wise GDP data
            table = soup.find("table", id="table_id")
            if table:
                # Initialize lists to store data
                ranks = []
                states = []
                gsdp_1819_current = []
                gsdp_1920_current = []
                share_1819 = []
                gdp_billion_dollars = []
                
                # Extract data from each row in the table
                rows = table.find_all("tr")[1:]  # Skip header row
                for row in rows:
                    columns = row.find_all("td")
                    rank = columns[0].text.strip()
                    state = columns[1].text.strip()
                    gsdp_1819 = columns[2].text.strip()
                    gsdp_1920 = columns[3].text.strip()
                    share = columns[4].text.strip()
                    gdp_billion = columns[5].text.strip()
                    
                    # Append data to respective lists
                    ranks.append(rank)
                    states.append(state)
                    gsdp_1819_current.append(gsdp_1819)
                    gsdp_1920_current.append(gsdp_1920)
                    share_1819.append(share)
                    gdp_billion_dollars.append(gdp_billion)
                
                # Print or process the data as needed
                for i in range(len(ranks)):
                    print(f"Rank: {ranks[i]}")
                    print(f"State: {states[i]}")
                    print(f"GSDP(18-19) at current prices: {gsdp_1819_current[i]}")
                    print(f"GSDP(19-20) at current prices: {gsdp_1920_current[i]}")
                    print(f"Share(18-19): {share_1819[i]}")
                    print(f"GDP ($ billion): {gdp_billion_dollars[i]}")
                    print()
            
            else:
                print("GDP table not found.")
        else:
            print(f"Failed to fetch GDP page. Status code: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"Error fetching GDP page: {str(e)}")

# Main function to orchestrate the scraping process
def main():
    economy_url = get_economy_url()
    if economy_url:
        print(f"Economy URL: {economy_url}")
        scrape_gdp_data(economy_url)
    else:
        print("Failed to get economy URL.")

if __name__ == "__main__":
    main()

```

```python
#  Scrape the details of trending repositories on Github.com. 
# Url = https://github.com/ 
# You have to find the following details: 
# A) Repository title 
# B) Repository description 
# C) Contributors count 
# D) Language used 
# pip install requests BeautifulSoup4


import requests
from bs4 import BeautifulSoup

# Function to scrape trending repositories data from GitHub homepage
def scrape_trending_repositories():
    url = "https://github.com/"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }

    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")

            # Find the section containing trending repositories
            trending_repos = soup.find_all("article", class_="Box-row")
            
            # Initialize lists to store data
            titles = []
            descriptions = []
            contributors_counts = []
            languages = []

            # Iterate through each trending repository
            for repo in trending_repos:
                title = repo.find("h1", class_="h3").text.strip()
                description = repo.find("p", class_="col-9").text.strip()
                contributors_count = repo.find("a", class_="Link--muted").text.strip()
                language = repo.find("span", itemprop="programmingLanguage").text.strip() if repo.find("span", itemprop="programmingLanguage") else "Not specified"

                # Append data to respective lists
                titles.append(title)
                descriptions.append(description)
                contributors_counts.append(contributors_count)
                languages.append(language)

            # Print or process the data as needed
            for i in range(len(titles)):
                print(f"Repository title: {titles[i]}")
                print(f"Repository description: {descriptions[i]}")
                print(f"Contributors count: {contributors_counts[i]}")
                print(f"Language used: {languages[i]}")
                print()

        else:
            print(f"Failed to fetch GitHub homepage. Status code: {response.status_code}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching GitHub homepage: {str(e)}")

# Main function to orchestrate the scraping process
def main():
    scrape_trending_repositories()

if __name__ == "__main__":
    main()


    
```

```python
# Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the 
# following details: 
# A) Song name 
# B) Artist name 
# C) Last week rank 
# D) Peak rank 
# E) Weeks on board 
#  Note: - From the home page you have to click on the charts option then hot 100-page link through code. 

import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# Function to initialize and configure Selenium webdriver
def init_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # To run Chrome in headless mode
    driver = webdriver.Chrome(executable_path="chromedriver", options=options)  # Change executable_path as needed
    return driver

# Function to scrape top 100 songs data from Billboard Hot 100 chart
def scrape_billboard_top100(driver):
    url = "https://www.billboard.com/"
    driver.get(url)

    try:
        # Wait until the charts option is visible and click it
        charts_link = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.LINK_TEXT, "CHARTS"))
        )
        charts_link.click()

        # Wait until the Hot 100 page link is visible and click it
        hot_100_link = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.LINK_TEXT, "Hot 100"))
        )
        hot_100_link.click()

        # Wait until the page loads and then fetch the HTML content
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "chart-list__element"))
        )

        # Parse the HTML using BeautifulSoup
        soup = BeautifulSoup(driver.page_source, "html.parser")

        # Initialize lists to store data
        song_names = []
        artist_names = []
        last_week_ranks = []
        peak_ranks = []
        weeks_on_board = []

        # Find all song elements
        songs = soup.find_all("li", class_="chart-list__element")

        # Iterate through each song element and extract data
        for song in songs:
            song_name = song.find("span", class_="chart-element__information__song").text.strip()
            artist_name = song.find("span", class_="chart-element__information__artist").text.strip()
            last_week_rank = song.find("span", class_="chart-element__meta text--center color--secondary text--last").text.strip()
            peak_rank = song.find("span", class_="chart-element__meta text--center color--secondary text--peak").text.strip()
            weeks_on_chart = song.find("span", class_="chart-element__meta text--center color--secondary text--week").text.strip()

            # Append data to respective lists
            song_names.append(song_name)
            artist_names.append(artist_name)
            last_week_ranks.append(last_week_rank)
            peak_ranks.append(peak_rank)
            weeks_on_board.append(weeks_on_chart)

        # Print or process the data as needed
        for i in range(len(song_names)):
            print(f"Song Name: {song_names[i]}")
            print(f"Artist Name: {artist_names[i]}")
            print(f"Last Week Rank: {last_week_ranks[i]}")
            print(f"Peak Rank: {peak_ranks[i]}")
            print(f"Weeks on Board: {weeks_on_board[i]}")
            print()

    finally:
        driver.quit()

# Main function to orchestrate the scraping process
def main():
    driver = init_driver()
    scrape_billboard_top100(driver)

if __name__ == "__main__":
    main()

```

```python
# Scrape the details of Highest selling novels. 
# A) Book name 
# B) Author name 
# C) Volumes sold 
# D) Publisher 
# E) Genre 
#  Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare
 

import requests
from bs4 import BeautifulSoup

url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"

def scrape_highest_selling_novels(url):
    try:
        # Send a GET request to fetch the HTML content
        response = requests.get(url)
        if response.status_code == 200:
            # Parse the HTML using BeautifulSoup
            soup = BeautifulSoup(response.content, "html.parser")

            # Find the table containing highest selling novels data
            table = soup.find("table", class_="in-article sortable")
            if table:
                # Initialize lists to store data
                books = []
                authors = []
                volumes_sold = []
                publishers = []
                genres = []

                # Extract data from each row in the table
                rows = table.find_all("tr")[1:]  # Skip header row
                for row in rows:
                    columns = row.find_all("td")
                    book = columns[1].text.strip()
                    author = columns[2].text.strip()
                    volumes = columns[3].text.strip()
                    publisher = columns[4].text.strip()
                    genre = columns[5].text.strip()

                    # Append data to respective lists
                    books.append(book)
                    authors.append(author)
                    volumes_sold.append(volumes)
                    publishers.append(publisher)
                    genres.append(genre)

                # Print or process the data as needed
                for i in range(len(books)):
                    print(f"Book Name: {books[i]}")
                    print(f"Author Name: {authors[i]}")
                    print(f"Volumes Sold: {volumes_sold[i]}")
                    print(f"Publisher: {publishers[i]}")
                    print(f"Genre: {genres[i]}")
                    print()

            else:
                print("Table not found on the page.")

        else:
            print(f"Failed to fetch URL: {url}. Status code: {response.status_code}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {url}. Error: {str(e)}")

# Main function to orchestrate the scraping process
def main():
    scrape_highest_selling_novels(url)

if __name__ == "__main__":
    main()

```

```python
# Scrape the details most watched tv series of all time from imdb.com. 
# Url = https://www.imdb.com/list/ls095964455/ You have 
# to find the following details: 
# A) Name 
# B) Year span 
# C) Genre 
# D) Run time 
# E) Ratings 
# F) Votes 

import requests
from bs4 import BeautifulSoup

url = "https://www.imdb.com/list/ls095964455/"

def scrape_most_watched_tv_series(url):
    try:
        # Send a GET request to fetch the HTML content
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            # Parse the HTML using BeautifulSoup
            soup = BeautifulSoup(response.content, "html.parser")

            # Find the list container
            list_container = soup.find("div", class_="lister-list")
            if list_container:
                # Find all list items (TV series)
                tv_series = list_container.find_all("div", class_="lister-item-content")

                # Initialize lists to store data
                names = []
                year_spans = []
                genres = []
                run_times = []
                ratings = []
                votes = []

                # Extract data from each TV series item
                for series in tv_series:
                    # Name
                    name = series.find("a").text.strip()
                    names.append(name)

                    # Year Span
                    year_span = series.find("span", class_="lister-item-year").text.strip()
                    year_spans.append(year_span)

                    # Genre
                    genre = series.find("span", class_="genre").text.strip()
                    genres.append(genre)

                    # Run Time
                    run_time = series.find("span", class_="runtime").text.strip()
                    run_times.append(run_time)

                    # Ratings
                    rating = series.find("div", class_="ipl-rating-star small").text.strip()
                    ratings.append(rating)

                    # Votes
                    vote = series.find("span", attrs={"name": "nv"}).text.strip()
                    votes.append(vote)

                # Print or process the data as needed
                for i in range(len(names)):
                    print(f"Name: {names[i]}")
                    print(f"Year Span: {year_spans[i]}")
                    print(f"Genre: {genres[i]}")
                    print(f"Run Time: {run_times[i]}")
                    print(f"Ratings: {ratings[i]}")
                    print(f"Votes: {votes[i]}")
                    print()

            else:
                print("List container not found on the page.")

        else:
            print(f"Failed to fetch URL: {url}. Status code: {response.status_code}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {url}. Error: {str(e)}")

# Main function to orchestrate the scraping process
def main():
    scrape_most_watched_tv_series(url)

if __name__ == "__main__":
    main()


    
```

```python
# Details of Datasets from UCI machine learning repositories. 
# Url = https://archive.ics.uci.edu/ You 
# have to find the following details: 
# A) Dataset name 
# B) Data type 
# C) Task 
# D) Attribute type 
# E) No of instances 
# F) No of attribute G) Year 
#  Note: - from the home page you have to go to the Show All Dataset page through code.

import requests
from bs4 import BeautifulSoup
import re

url = "https://archive.ics.uci.edu/"

def scrape_datasets_from_uci(url):
    try:
        # Send a GET request to fetch the home page
        response = requests.get(url)
        if response.status_code == 200:
            # Parse the HTML using BeautifulSoup
            soup = BeautifulSoup(response.content, "html.parser")

            # Find and click on the link to show all datasets
            link = soup.find("a", href=re.compile(r"datasets\.php$"))
            if link:
                datasets_url = url + link.get("href")
                datasets_response = requests.get(datasets_url)
                if datasets_response.status_code == 200:
                    # Parse the datasets page using BeautifulSoup
                    datasets_soup = BeautifulSoup(datasets_response.content, "html.parser")

                    # Find the table containing datasets
                    table = datasets_soup.find("table", {"border": "1", "cellpadding": "5"})

                    if table:
                        # Initialize lists to store data
                        dataset_names = []
                        data_types = []
                        tasks = []
                        attribute_types = []
                        instances = []
                        attributes = []
                        years = []

                        # Extract data from each row in the table
                        rows = table.find_all("tr")[1:]  # Skip header row
                        for row in rows:
                            columns = row.find_all("td")

                            # Dataset Name
                            dataset_name = columns[0].text.strip()
                            dataset_names.append(dataset_name)

                            # Data Type
                            data_type = columns[1].text.strip()
                            data_types.append(data_type)

                            # Task
                            task = columns[2].text.strip()
                            tasks.append(task)

                            # Attribute Type
                            attribute_type = columns[3].text.strip()
                            attribute_types.append(attribute_type)

                            # No of Instances
                            instances_count = columns[4].text.strip()
                            instances.append(instances_count)

                            # No of Attributes
                            attributes_count = columns[5].text.strip()
                            attributes.append(attributes_count)

                            # Year
                            year = columns[6].text.strip()
                            years.append(year)

                        # Print or process the data as needed
                        for i in range(len(dataset_names)):
                            print(f"Dataset Name: {dataset_names[i]}")
                            print(f"Data Type: {data_types[i]}")
                            print(f"Task: {tasks[i]}")
                            print(f"Attribute Type: {attribute_types[i]}")
                            print(f"No of Instances: {instances[i]}")
                            print(f"No of Attributes: {attributes[i]}")
                            print(f"Year: {years[i]}")
                            print()

                    else:
                        print("Table not found on the datasets page.")

                else:
                    print(f"Failed to fetch datasets URL: {datasets_url}. Status code: {datasets_response.status_code}")

            else:
                print("Link to show all datasets not found on the home page.")

        else:
            print(f"Failed to fetch URL: {url}. Status code: {response.status_code}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {url}. Error: {str(e)}")

# Main function to orchestrate the scraping process
def main():
    scrape_datasets_from_uci(url)

if __name__ == "__main__":
    main()


    
```
